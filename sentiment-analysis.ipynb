```python
import pandas as pd


file_path = "/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv"


df = pd.read_csv(file_path)


print(df.head())
print(df.info())
```

```python
import re
import string

def cleantext(text):
    if text is None:
        return ""
    text = str(text).lower()                       
    text = re.sub(r"<.*?>", "", text)             
    text = re.sub(r"http\S+|www\S+", "", text)     
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)  
    return text.strip()
```

```python
df['clean_review']=df['review'].apply(cleantext)
df['sentiment']=df['sentiment'].map({"positive":0,"negative":1})
print(df['clean_review'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['clean_review'],df['sentiment'],test_size=0.2,random_state=42)
```

```python
x = df['clean_review'].values 
print(x)
y = df['sentiment'].values 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

c = TfidfVectorizer(ngram_range=(1,2),max_features=10000)
x_train_vec = c.fit_transform(x_train)  # fit only on training text
x_test_vec = c.transform(x_test)        # transform test text
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
clf=LogisticRegression(max_iter=500,C=1.0)
clf.fit(x_train_vec,y_train)
y_pred=clf.predict(x_test_vec)
print(accuracy_score(y_pred,y_test))
```
